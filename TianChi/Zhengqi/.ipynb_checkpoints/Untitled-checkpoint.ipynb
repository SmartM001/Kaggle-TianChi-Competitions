## 数据读取

```python
import numpy as np
import pandas as pd

data_tra = pd.read_csv("data\\zhengqi_train.txt", sep='\t')
data_tes = pd.read_csv("data\\zhengqi_test.txt", sep='\t')
train = data_tra
train_x = train.drop(['target'], axis=1)
data_all = pd.concat([train_x, data_tes])
```

## 数据分析与数据处理

```python
import seaborn as sea
import matplotlib.pyplot as plt

# 查看每个特征的数据分布
for col in data_all.columns:
    sea.distplot(data_tra[col])
    sea.distplot(data_tes[col])
    plt.show()
```

#### 从上面的数据分布看，V5,V9,V11,V17,V22,V28的训练集和测试集合分布不一致，会导致模型泛化能力差，采取删除这一部分特征

## 删除无效特征，并做标准化

```python
from sklearn.preprocessing import StandardScaler

data_all.drop(['V5','V9','V11','V17','V22','V28'], axis=1, inplace=True)

scaler = StandardScaler()
data_all_std = pd.DataFrame(scaler.fit_transform(data_all), columns=data_all.columns)
```

```python
for col in data_all_std.columns:
    sea.distplot(data_all_std[col])
    sea.distplot(data_tra[col])
    sea.distplot(data_tes[col])
    plt.show()
```

## 切分数据集

```python
x_train = data_all_std.ix[0 : len(data_tra)-1]
x_test = data_all_std.ix[len(data_tra) : ]
y_train = data_tra['target']
```

```python
sea.distplot(y_train)
```

## 去掉方差较小的特征

```python
from sklearn.feature_selection import VarianceThreshold

threshold = 0.85
vt = VarianceThreshold().fit(x_train)
feature_var_threshold = x_train.columns[vt.variances_ > threshold*(1-threshold)]

x_train_vt = x_train[feat_var_threshold]
x_test_vt = x_test[feat_var_threshold]
```

## 单变量选择

```python
from sklearn.feature_selection import f_regression
from sklearn.feature_selection import SelectKBest

# 对去掉方差最小特征后的数据做单变量选择
x_score = SelectKBest(score_func=f_regression, k='all').fit(x_train_vt, y_train)
feature_score = pd.DataFrame({"feature": x_train_vt.columns, "score":x_score.scores_})
# 选择得分TOP19的特征，得分100以上
feature = feature_score.sort_values('score', ascending=False).head(19)['feature']
x_train_vt_fs = x_train_vt[feature]
x_test_vt_fs = x_test_vt[feature]
```

## 模型选择

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, BaggingRegressor
from sklearn.model_selection import KFold, cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.neural_network import MLPRegressor
from sklearn.pipeline import make_pipeline
from lightgbm import LGBMRegressor
from xgboost import XGBRegressor
from sklearn.svm import SVR

n_folds = 10
def rsmle_cv(model, x_train):
    kf = KFold(n_folds, shuffle=True, random_state=42).get_n_splits(x_train)
    rmse= -cross_val_score(model, x_train, y_train, scoring="neg_mean_squared_error", cv = kf)
    return rmse

clf_linear = make_pipeline(LinearRegression())
clf_lasso = make_pipeline(Lasso(alpha=0.005))
clf_ridge = make_pipeline(Ridge(alpha=0.012))
clf_elasticNet = make_pipeline(ElasticNet(alpha=0.005, l1_ratio=0.9))
clf_randFroest = make_pipeline(RandomForestRegressor())
clf_svr = make_pipeline(SVR(kernel='linear'))
clf_mlpr = make_pipeline(MLPRegressor())

clf_xgb = make_pipeline(XGBRegressor(booster='gbtree',colsample_bytree=0.8, gamma=0.1, 
                             learning_rate=0.02, max_depth=5, 
                             n_estimators=500,min_child_weight=0.8,
                             reg_alpha=0, reg_lambda=1,
                             subsample=0.8, silent=1,
                             random_state =42, nthread = 2))

clf_lgb = make_pipeline(LGBMRegressor(
        learning_rate=0.01,
        max_depth=-1,
        n_estimators=5000,
        boosting_type='gbdt',
        random_state=2018,
        objective='regression',))


score = rsmle_cv(clf_svr, x_train_vt_fs)
print("SVR 得分: {:.4f} ({:.4f})\n".format(score.mean(), score.std()))
score = rsmle_cv(clf_linear, x_train_vt_fs)
print("linear 得分: {:.4f} ({:.4f})\n".format(score.mean(), score.std()))
score = rsmle_cv(clf_lasso, x_train_vt_fs)
print("Lasso 得分: {:.4f} ({:.4f})\n".format(score.mean(), score.std()))
score = rsmle_cv(clf_ridge, x_train_vt_fs)
print("Ridge 得分: {:.4f} ({:.4f})\n".format(score.mean(), score.std()))
score = rsmle_cv(clf_elasticNet, x_train_vt_fs)
print("ElasticNet 得分: {:.4f} ({:.4f})\n".format(score.mean(), score.std()))
score = rsmle_cv(clf_randFroest, x_train_vt_fs)
print("RandomForest 得分: {:.4f} ({:.4f})\n".format(score.mean(), score.std()))
score = rsmle_cv(clf_mlpr, x_train_vt_fs)
print("MLP 得分: {:.4f} ({:.4f})\n".format(score.mean(), score.std()))
score = rsmle_cv(clf_xgb, x_train_vt_fs)
print("XGB 得分: {:.4f} ({:.4f})\n".format(score.mean(), score.std()))
score = rsmle_cv(clf_lgb, x_train_vt_fs)
print("LGB 得分: {:.4f} ({:.4f})\n".format(score.mean(), score.std()))
```

```python
clf_lasso.fit(x_train_vt_fs, y_train)
clf_elasticNet.fit(x_train_vt_fs, y_train)
clf_mlpr.fit(x_train_vt_fs, y_train)
```

```python
from sklearn.base import BaseEstimator, TransformerMixin, RegressorMixin, clone

#简单模型融合
class AveragingModels(BaseEstimator, RegressorMixin, TransformerMixin):
    def __init__(self, models):
        self.models = models
        
    # 遍历所有模型
    def fit(self, X, y):
        self.models_ = [clone(x) for x in self.models]
        
        for model in self.models_:
            model.fit(X, y)

        return self
    
    # 预估，并对预估结果值做average
    def predict(self, X):
        predictions = np.column_stack([
            model.predict(X) for model in self.models_
        ])
    
        return np.mean(predictions, axis=1)   
    
```

```python
averaged_models = AveragingModels(models = (clf_lasso,clf_elasticNet,clf_mlpr))

score = rsmle_cv(averaged_models, x_train_vt_fs)
print(" 对基模型集成后的得分: {:.4f} ({:.4f})\n".format(score.mean(), score.std()))
```

```python
averaged_models.fit(x_train_vt_fs, y_train)
```

```python
y_pre = averaged_models.predict(x_test_vt_fs)
np.savetxt("zhengqi_predict.txt", y_pre)
```
